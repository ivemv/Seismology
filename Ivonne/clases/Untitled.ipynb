```python
import obspy
```

```python
from obspy.clients.fdsn import Client
from obspy import UTCDateTime
```

```python
t = UTCDateTime(2017,1,1,0,0,0)
```

```python
t2 = UTCDateTime(2018,5,1,0,0,0)
```

```python
client=Client("IRIS")
```

```python
catalog = client.get_events(starttime=t, endtime=t2,
                            minmagnitude=6.5)
```

```python
print(catalog)
```

```python
catalog.plot();
```

```python
cli = Client("ISC")
```

```python
cat2 = cli.get_events(starttime=t,endtime=t2, minmagnitude=6.5)
```

```python
print(cat2)
```

```python
cat2.plot();
```
