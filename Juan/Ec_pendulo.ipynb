#### Ecuación del péndulo

*Para el caso de un péndulo vertical el cual está conformado por una masa suspendida a un punto por un material (ej. un hilo).

La masa del péndulo vertical está sometida a dos fuerzas diferentes, la tensión del material que lo sostiene y la fuerza de gravedad que puede ser descompuesta en una componente tangencial y normal al movimiento de la masa. Tanto la tensión como la componente normal son de magnitudes iguales pero direcciones diferentes por lo que se anulan. 

![Esquema de fuerzas para el péndulo vertical](/Pendulo.jpg)


La fuerza que describe el movimiento oscilante de la masa es pues la componente tangencial de la gravedad que puede ser descrita de acuerdo a la geometría del pendulo e igualada a la segunda ley de Newton para obtener la ecuación de movimiento:  

$F_t = mg\sin \theta = ma$  ( Ec. 1)

Es desplazamiento está dado por la longitud de la cuerda por el cambio del ángulo entre la cuerda y la vertical en el tiempo; además sabiendo que la aceleración es la doble derivada del desplazamiento se deduce que:

$a = l \frac{d^2 \theta}{dt^2} $  ( Ec. 2)

Sustituyendo Ec. 2 en Ec. 1:

$ -mg\sin \theta = m l \frac{d^2 \theta}{dt^2} $    (Ec. 3)

Agregando un signo negativo debido a que el ángulo $\theta$ disminuye. Reduciendo términos en Ec. 3 y reorganizando:

$\frac{d^2 \theta}{dt^2} = -\frac{g}{l} \sin \theta $  (Ec. 4)

Esta ecuación diferencial de segundo orden tiene la solución:

$\theta = A(\omega t + \phi_0) $

Donde A es la amplitud de la oscilación, $\phi_0$ es la el angulo de desfasamiento y $\omega$ es la frecuencia angular de oscilación.

$\omega = \sqrt{\frac{l}{g}} $

Recordando que el periodo de oscilación (T) es inverso a la frecuencia podemos obtener que:

$ T = \frac{2 \pi}{\omega} = {2 \pi}\sqrt{\frac{l}{g}}$


De esta última relación es posible darnos cuenta que el periodo de oscilación de un pendulo no depende de la masa ni la amplitud inicial con la que inicia el movimiento, únicamente depende de la longitud de la cuerda y de la gravedad.







### Tipos de sismómetros

##### Periodo Corto (< 2s)
Su uso principal es para estudiar ondas de cuerpo (P y S) generadas por sismicidad local, regional y explosiones nucleares cercanas (distancias menores a 500 km). [1]
	

##### Periodo Largo
Permite estudiar tanto ondas de cuerpo como superficiales, así como las oscilaciones libres de la Tierra y las mareas terrestres. [1]

##### Banda ancha
Estos sismometros tiene una sensibilidad alta en amplio rango de de frecuencias. Pueden ser utilizadas para observar señales telesísmicas, construir mecanismos focales con gran precisión y registrar miscrosismicidad. [2]
