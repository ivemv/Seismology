#### Ecuacin del pndulo

*Para el caso de un pndulo vertical el cual est conformado por una masa suspendida a un punto por un material (ej. un hilo).

La masa del pndulo vertical est sometida a dos fuerzas diferentes, la tensin del material que lo sostiene y la fuerza de gravedad que puede ser descompuesta en una componente tangencial y normal al movimiento de la masa. Tanto la tensin como la componente normal son de magnitudes iguales pero direcciones diferentes por lo que se anulan. 

![Esquema de fuerzas para el pndulo vertical](/Pendulo.jpg)


La fuerza que describe el movimiento oscilante de la masa es pues la componente tangencial de la gravedad que puede ser descrita de acuerdo a la geometra del pendulo e igualada a la segunda ley de Newton para obtener la ecuacin de movimiento:  

$F_t = mg\sin \theta = ma$  ( Ec. 1)

Es desplazamiento est dado por la longitud de la cuerda por el cambio del ngulo entre la cuerda y la vertical en el tiempo; adems sabiendo que la aceleracin es la doble derivada del desplazamiento se deduce que:

$a = l \frac{d^2 \theta}{dt^2} $  ( Ec. 2)

Sustituyendo Ec. 2 en Ec. 1:

$ -mg\sin \theta = m l \frac{d^2 \theta}{dt^2} $    (Ec. 3)

Agregando un signo negativo debido a que el ngulo $\theta$ disminuye. Reduciendo trminos en Ec. 3 y reorganizando:

$\frac{d^2 \theta}{dt^2} = -\frac{g}{l} \sin \theta $  (Ec. 4)

Esta ecuacin diferencial de segundo orden tiene la solucin:

$\theta = A(\omega t + \phi_0) $

Donde A es la amplitud de la oscilacin, $\phi_0$ es la el angulo de desfasamiento y $\omega$ es la frecuencia angular de oscilacin.

$\omega = \sqrt{\frac{l}{g}} $

Recordando que el periodo de oscilacin (T) es inverso a la frecuencia podemos obtener que:

$ T = \frac{2 \pi}{\omega} = {2 \pi}\sqrt{\frac{l}{g}}$


De esta ltima relacin es posible darnos cuenta que el periodo de oscilacin de un pendulo no depende de la masa ni la amplitud inicial con la que inicia el movimiento, nicamente depende de la longitud de la cuerda y de la gravedad.







### Tipos de sismmetros

##### Periodo Corto (< 2s)
Su uso principal es para estudiar ondas de cuerpo (P y S) generadas por sismicidad local, regional y explosiones nucleares cercanas (distancias menores a 500 km). [1]
	

##### Periodo Largo
Permite estudiar tanto ondas de cuerpo como superficiales, as como las oscilaciones libres de la Tierra y las mareas terrestres. [1]

##### Banda ancha
Estos sismometros tiene una sensibilidad alta en amplio rango de de frecuencias. Pueden ser utilizadas para observar seales telessmicas, construir mecanismos focales con gran precisin y registrar miscrosismicidad. [2]
