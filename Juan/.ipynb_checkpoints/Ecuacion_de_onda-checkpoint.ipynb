# Ecuación de onda


### Formulación para un medio isotrópico
Partiendo de la ecuación de la ley de Hooke 
\begin{equation}
\sigma_{ij}=[\lambda\delta_{ij}\epsilon_{kk} + 2\mu\epsilon_{ij}]=\lambda\delta_{ij}\theta + 2\mu\epsilon_{ij}
\end{equation}

(Donde $\mu$ representa la resistencia al corte del material y $\theta$ mide la dilatancia o el cambio del volumen del cuerpo, definida como la traza del tensor $\epsilon_{ij}$, que es la suma de las componentes cuyos subíndices son iguales)

y sustituyendola en la segunda ley de Newton podemos obtener la ecuación de movimiento: 

\begin{equation}
 \rho \frac{\delta^2 u_{i}}{\delta t^2} = f_{i} + \frac{\delta \sigma_{ij}}{\delta x_{j}}
\end{equation}

Donde la masa la podemos expresar en términos de la densidad y dado que los volúmenes manejados son infinitesimales los podemos despreciar; además $f_{i}$ representa fuerzas externas que para este análisis se toman como cero. Reduciendo y asumiendo que las constantes de Lamé son ctes. y sustituyendo $\sigma_{ij}$ (en este caso $\sigma_{11}$) obtenemos la ecuación (7):
 \begin{equation}
 \rho \frac{\delta^{2}u_{1}}{\delta t^{2}} = \lambda \frac{\delta \theta}{\delta x^{1}}+ \mu \frac{\delta}{\delta x^{1}}(\frac{\delta u_{1}}{\delta x_{1}} + \frac{\delta u_{2}}{\delta x_{2}} + \frac{\delta u_{3}}{\delta x_{3}}) + \mu (\frac{\delta^{2}u_{1}}{\delta x^{2}_{1}} + \frac{\delta^{2}u_{1}}{\delta x^{2}_{2}} + \frac{\delta^{2}u_{1}}{\delta x^{2}_{3}})
 \end{equation}
Las ecuaciones para las demas componentes de $\sigma_{ij}$ tienen forma similar a la ecuación 7, donde i es el subíndice de la dirección de moviemiento u y j es la el subíndice de la derivada (la dirección del cambio).

Las sumas de las derivadas parciales se pueden expresar como la divergencia del vector derivado (en este caso: u). 

Las sumas de las dobles derivadas parciales corresponden al laplaciano del vector.


Así podemos reducir y reescribir la ecuación:

\begin{equation}
\rho \frac{\delta^{2}u}{\delta t^{2}}= (\lambda + \mu) \nabla (\nabla . u) + \mu \nabla^{2} \frac{\delta^{2}u}{\delta x^{2}}
\end{equation}

Por la identidad $\nabla^{2}u =\nabla (\nabla . u) - (\nabla \times \nabla \times u) $

La ecuación queda: 

\begin{equation}
\rho \frac{\delta^{2}u}{\delta t^{2}}= (\lambda + 2\mu) \nabla (\nabla . u) - (\mu \nabla \times \nabla \times u)
\end{equation}

El teorema de Helmhotz nos indica que podemos representar el desplazamiento como la suma de un campo escalar $(\Phi)$ sin rotación y un campo vectorial $(\Psi)$ sin divergencia:

\begin{eqnarray}
u=\nabla . \Phi + \nabla \times \psi \\
(\nabla \times \Phi ) = 0 \\
(\nabla . \Psi ) = 0
\end{eqnarray}

Sustituyendo en la ecuación (9) se obtiene: 
\begin{equation}
\nabla[(\lambda + 2 \mu) \nabla^{2} \Phi - \rho \frac{\delta^{2}\Phi}{\delta t^{2}}] + \nabla \times [\mu \nabla^{2}\Psi - \rho \frac{\delta^{2}\Psi}{\delta t^{2}}] = 0
\end{equation}

Separando esta ecuación en sus componentes escalar y rotacional, definiendo dos parámetros, $\alpha$ y $\beta$ como: \\
\begin{eqnarray}
\alpha = \sqrt[]{\frac{\lambda + 2 \mu}{\rho}}\\
\beta = \sqrt[]{\frac{\mu}{\rho}}
\end{eqnarray}
\\

Podemos obtener las soluciones de la ecuación de onda para el campo escalar $(\Phi)$ que representa una onda de cambio de volumen correspondiente a las ondas P, y la ecuación para el campo rotacional $(\Psi)$ que contempla únicamente deformación tangencial, denominada onda S: 

\begin{eqnarray}
\frac{\delta^2 \Phi}{\delta x^2} - \frac{1}{\alpha^2}\frac{\delta^2 \Phi}{\delta t^2} = 0 \\
\frac{\delta^2 \Psi}{\delta x^2} - \frac{1}{\beta^2}\frac{\delta^2 \Psi}{\delta t^2} = 0
\end{eqnarray}

### Ecuación de ondas en una dimensión
Para la transmisión de ondas en una sola dimensión y basándonos en que de la ecuación constitutiva se obtiene que $\sigma_{ij}= E\epsilon_{ij}$, donde E es el módulo de Young, que a su vez podemos calcular mediante otros módulos elásticos, la ecuación de movimiento puede escribirse como:

\begin{equation}
E \frac{\delta^2 u_{i}}{\delta x_{i}^2} = \rho \frac{\delta^2 u_{i}}{\delta t^2}
\end{equation}

La forma general de la ecuación de onda dependiente de la posición y el tiempo es:

\begin{equation}
\frac{\delta^2 u_{i}}{\delta x_{i}^2} = \frac{1}{c^2} \frac{\delta^2 u_{i}}{\delta t^2}
\end{equation}

Donde c es la velocidad de transmisión de las ondas. La solución general, propuesta por Jean D’Alambert, es:

$u_{1}(x_{1},t) = f(x_{1}- ct) + g(x_{1} + ct)$\\

Donde f y g son funciones arbitrarias que satisfacen las condiciones iniciales de la propagación de la onda.

Con las ecuaciones anteriores deducimos que el factor de transmisión de ondas es:

$ c = (E /\rho)^{1/2} $

Un aumento en la densidad tiende a reducir la velocidad, sin embargo, el módulo de Young tiende a aumentar con la densidad causando un efecto compensador que usualmente da un aumento en el valor de c, por lo que el aumento de la densidad de un material viene acompañado de un aumento en la velocidad de propagación de ondas.



### Solución armónica 

Una solución común para la ecuación de onda es   
\begin{equation}
u(x,t) = A sin [(kx − ct) + φ]
\end{equation}
donde A es la amplitud, k el número de onda y φ la fase inicial (todas ellas constantes y determinadas por las condiciones iniciales).

En esta clase de soluciones se definen varias propiedades que describen el movimiento de la onda, la longitud de onda λ, la frecuencia angular ω, la frecuencia f y el periodo T:

$ λ =2π/k$   ;   

$ω = kc = 2πf = 2π/T $

Dado que la frecuencia angular está relacionada con la velocidad en muchas ocuaciones la solución se escribe tomando en cuenta ω:
\begin{equation}
u(x,t) = A sin(kx − ωt + φ)
\end{equation}
